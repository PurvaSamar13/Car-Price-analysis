```python
D = data.groupby('year').selling_price.sum()
D
```

```python
x = D.index.values.reshape(-1,1)
x
```

```python
y = D.values
y
```

```python
plt.scatter(x,y)
plt.xlabel('Year')
plt.ylabel('Selling price')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y
```

```python
Yp = reg.predict(x)
Yp
```

```python
plt.scatter(x,y)
plt.plot(x,Yp)
plt.show()
```

```python
reg.score(x,y)*100
```

```python
reg.predict([[2025]])
```
